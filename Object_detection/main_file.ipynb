## Setup Folders

```python
# get current file location 
current_directory = os.getcwd()
# create a new folder datasets
datasets_directory = os.path.join(current_directory, 'datasets')
if not os.path.exists(datasets_directory):
    os.makedirs(datasets_directory)
    print(f"Dataset directory created @ {datasets_directory}")
else:
    print(f"Directory already exists @ {datasets_directory}")
# set environment variable name to the folder location 
# THIS IS WHERE THE DATASET WILL BE DOWNLOADED 
os.environ["DATASET_DIRECTORY"] = datasets_directory 

```
